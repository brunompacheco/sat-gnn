```python
import pandas as pd 
import wandb
```

```python
api = wandb.Api()

# Project is specified by <entity/project-name>
runs = api.runs("brunompac/sat-gnn", filters={'group': 'GridSearch-MultiTarget'})

summary_list, config_list, name_list = [], [], []
for run in runs: 
    # .summary contains the output keys/values for metrics like accuracy.
    #  We call ._json_dict to omit large files 
    summary_list.append(run.summary._json_dict)

    # .config contains the hyperparameters.
    #  We remove special values that start with _.
    config_list.append(
        {k: v for k,v in run.config.items()
          if not k.startswith('_')})

    # .name is the human-readable name of the run.
    name_list.append({'name': run.name, 'id': run.id})

# df = pd.DataFrame({
#     "summary": summary_list,
#     "config": config_list,
#     "name": name_list,
# })
df = pd.concat([pd.DataFrame(summary_list), pd.DataFrame(config_list), pd.DataFrame(name_list)], axis=1)
df.drop(columns=[
    'val_time_backward', 'train_time_backward', 'train_loss', 'val_time_loss',
    '_step','val_time', '_timestamp', 'train_time', 'val_loss_mean_gap',
    'train_time_loss', 'val_time_forward', 'val_loss_accuracy',
    'train_time_forward','_wandb', 'model', 'device', 'epochs','loss_func',
    'optimizer','random_seed','val_dataset', 'lr_scheduler', 'test_dataset',
    'train_dataset', 'mixed_precision', 'lr_scheduler_params',
], inplace=True)

df.sort_values(by='best_val').head(10)
```
